```python
print("\nEnter the Row Size of First Matrix: ", end="")
try:
    rOne = int(input())
    print("\nEnter the Column Size of First Matrix: ", end="")
    try:
        cOne = int(input("\n"))
        print("\nEnter " +str(rOne * cOne)+ " Elements: ", end="")
        mOne = []
        for i in range(rOne):
            mOne.append([])
            for j in range(cOne):
                try:
                    num = int(input("\n"))
                    mOne[i].append(num)
                except ValueError:
                    print("\nInvalid Input!")
                    exit()
        print("\nEnter Row Size of Second Matrix: ", end="")
        try:
            rTwo = int(input("\n"))
            if cOne == rTwo:
                print("\nEnter Column Size of Second Matrix: ", end="")
                try:
                    cTwo = int(input("\n"))
                    print("\nEnter " +str(rTwo * cTwo)+ " Elements: ", end="")
                    mTwo = []
                    for i in range(rTwo):
                        mTwo.append([])
                        for j in range(cTwo):
                            try:
                                num = int(input("\n"))
                                mTwo[i].append(num)
                            except ValueError:
                                print("\nInvalid Input!")
                                exit()

                    # now multiplying two matrices
                    mThree = []
                    for i in range(rOne):
                        mThree.append([])
                        for j in range(cTwo):
                            sum = 0
                            for k in range(cOne):
                                sum = sum + (mOne[i][k] * mTwo[k][j])
                            mThree[i].append(sum)

                    # now printing the multiplication result
                    print("\nMultiplication Result of Two Given Matrix is:")
                    for i in range(rOne):
                        for j in range(cTwo):
                            print(mThree[i][j], end=" ")
                        print()
                except ValueError:
                    print("\nInvalid Column Size!")
            else:
                print("\nMultiplication not possible!")
        except ValueError:
            print("\nInvalid Row Size!")
    except ValueError:
        print("\nInvalid Column Size!")
except ValueError:
    print("\nInvalid Row Size!")
```

Matrix multiplication is a binary operation that produces another matrix by multiplying two matrices. The elements of the matrix are multiplied using basic arithmetic. When two matrices are multiplied, the row elements of the first matrix are multiplied by the column elements of the second matrix.

Note: It is only possible to multiply two matrices if the number of columns in the first matrix equals the number of rows in the second matrix.

A matrix can be implemented as a hierarchical list in Python (list inside a list). Each element can be thought of as a row in the matrix.

A 3Ã—2 matrix, for example, would be represented by X = [[1, 2], [4, 5], [3, 6]].

X[0] can be used to choose the first row. Furthermore, the element in the first row and the first column can be chosen as X[0][0].

Matrix multiplication can only take place between compatible matrices and also that it is not commutative in nature.

Matrix multiplication in Python can also be done much more efficiently using numpy library methods, namely:
numpy.matmul()
numpy.dot()
numpy.multiply()
These methods are more efficient since they make use of vectorization, which makes their execution much faster than executing explicit for-loops in Python.
